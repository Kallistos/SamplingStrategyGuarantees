```python
import numpy as np
import matplotlib as mpl

def colorFader(c1,c2,mix=0): #fade (linear interpolate) from color c1 (at mix=0) to c2 (mix=1)
    c1=np.array(mpl.colors.to_rgb(c1))
    c2=np.array(mpl.colors.to_rgb(c2))
    return mpl.colors.to_hex((1-mix)*c1 + mix*c2)

old_df_fig4 = pd.read_csv('./data/sampling/area_aggregation.csv')

potential_area_df = pd.read_csv('./data/sampling/potential_area.csv')

green = '#00FF00'
red = '#FF0000'
black = '#000000'

min_val = -0.9
mid_val = 0.0
max_val = 0.5

def color_rules(val):
    if isinstance(val, (float, int)):
        if val > mid_val:
            pcolor = max(min(100.0*(val - mid_val)/(max_val - mid_val),100.0),0.00)
            color = colorFader(black, green, pcolor/100)
        else:
            pcolor = max(min(100.0*(mid_val - val)/(mid_val - min_val),100.0),0.00)
            color = colorFader(black, red, pcolor/100)
        return f"color: {color}"
    return ""

relative_areas = {
    'Sampling Strategy': [],
    'Subject System': [],
}
for sample_size in old_df_fig4['Sample Set Size'].unique():
    relative_areas['{}'.format(sample_size)] = []

# 7z case study
for strat in sampling_strategies:

    for cs in case_studies:
        df_new_fig_4 = old_df_fig4[(old_df_fig4['Case Study'] == cs) & (old_df_fig4['Sampling Strategy'] == strat)]

        relative_areas['Subject System'].append(subject_systems[cs])
        relative_areas['Sampling Strategy'].append(strat)
        for sample_size in df_new_fig_4['Sample Set Size'].unique():
            achieved_area = df_new_fig_4[df_new_fig_4['Sample Set Size'] == sample_size]['Area'].values[0]
            max_area = potential_area_df[potential_area_df['Sample Set Size'] == sample_size]['Max Area'].values[0]

            relative_areas['{}'.format(sample_size)].append(round(achieved_area / max_area, 3))

relative_areas_df = pd.DataFrame(relative_areas)
relative_areas_df.to_csv('./data/tikz/area_aggregation_{}_{}.csv'.format(cs, strat), index=False)
# Filter all columns that are not in the following array
relative_areas_df = relative_areas_df[['Sampling Strategy', 'Subject System', '10', '50', '100', '150', '200', '250', '300']]

negative_subset = relative_areas_df.loc[(relative_areas_df[relative_areas_df.select_dtypes(include='number').columns] < 0).all(1), relative_areas_df.select_dtypes(include='number').columns]

# print(relative_areas_df)

s = relative_areas_df.style.format_index(axis=1, escape="latex").applymap(color_rules).hide(axis="index")
s = s.format(formatter={"name": str.upper}, precision=3)

s.to_latex('../FSE25/graphics/rq2_table.tex', convert_css=True, hrules=True)
```
